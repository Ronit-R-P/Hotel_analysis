```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure the date column is in datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')

# Extract month and year
df['month_year'] = df['reservation_status_date'].dt.to_period('M')

# Filter data for 2016 only
filtered_df_2016 = df[df['reservation_status_date'].dt.year == 2016]

# Group by month_year, then calculate mean ADR
monthly_adr_2016 = filtered_df_2016.groupby('month_year')['adr'].mean().reset_index()

# Further filter for cancelled reservations in 2016
cancelled_df_2016 = filtered_df_2016[filtered_df_2016['is_canceled'] == 1]

# Group by month_year, then count cancellations
monthly_cancellations_2016 = cancelled_df_2016.groupby('month_year').size().reset_index(name='cancellations')

# Merge the two datasets on month_year
combined_data_adr_cancellations = pd.merge(monthly_adr_2016, monthly_cancellations_2016, on='month_year')

# Plot combined graph
fig, ax1 = plt.subplots(figsize=(20, 10))

# Plot mean ADR
color = 'tab:blue'
ax1.set_xlabel('Month-Year', fontsize=20)
ax1.set_ylabel('Mean ADR', fontsize=20, color=color)
ax1.plot(combined_data_adr_cancellations['month_year'].astype(str), combined_data_adr_cancellations['adr'], marker='o', color=color, linestyle='-', label='Mean ADR')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis to plot cancellations
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Number of Cancellations', fontsize=20, color=color)
ax2.plot(combined_data_adr_cancellations['month_year'].astype(str), combined_data_adr_cancellations['cancellations'], marker='x', color=color, linestyle='-', label='Cancellations')
ax2.tick_params(axis='y', labelcolor=color)

# Add legends
fig.tight_layout()
ax1.legend(loc='upper left', fontsize=15)
ax2.legend(loc='upper right', fontsize=15)
plt.title('Mean ADR and Cancellations by Month in 2016', fontsize=25)
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure the date column is in datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')

# Extract month and year
df['month_year'] = df['reservation_status_date'].dt.to_period('M')

# Filter data for 2016 only
filtered_df_2016 = df[df['reservation_status_date'].dt.year == 2016]

# Group by month_year, then calculate mean ADR
monthly_adr_2016 = filtered_df_2016.groupby('month_year')['adr'].mean().reset_index()

# Group by month_year, then count total bookings
monthly_bookings_2016 = filtered_df_2016.groupby('month_year').size().reset_index(name='bookings')

# Merge the two datasets on month_year
combined_data_adr_bookings = pd.merge(monthly_adr_2016, monthly_bookings_2016, on='month_year')

# Plot combined graph
fig, ax1 = plt.subplots(figsize=(20, 10))

# Plot mean ADR
color = 'tab:red'
ax1.set_xlabel('Month-Year', fontsize=20)
ax1.set_ylabel('Mean ADR', fontsize=20, color=color)
ax1.plot(combined_data_adr_bookings['month_year'].astype(str), combined_data_adr_bookings['adr'], marker='o', color=color, linestyle='-', label='Mean ADR')
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis to plot total bookings
ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Number of Bookings', fontsize=20, color=color)
ax2.plot(combined_data_adr_bookings['month_year'].astype(str), combined_data_adr_bookings['bookings'], marker='x', color=color, linestyle='-', label='Total Bookings')
ax2.tick_params(axis='y', labelcolor=color)

# Add legends
fig.tight_layout()
ax1.legend(loc='upper left', fontsize=15)
ax2.legend(loc='upper right', fontsize=15)
plt.title('Mean ADR and Total Bookings by Month in 2016', fontsize=25)
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd

# Ensure the date column is in datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')

# Extract month and year
df['month_year'] = df['reservation_status_date'].dt.to_period('M')

# Group by month_year, then calculate mean ADR and total reservations
monthly_adr_reservations = df.groupby('month_year').agg({'adr': 'mean', 'reservation_status_date': 'size'}).rename(columns={'reservation_status_date': 'reservations'}).reset_index()

# Compute the correlation between ADR and total reservations
correlation = monthly_adr_reservations['adr'].corr(monthly_adr_reservations['reservations'])

# Display the correlation coefficient
print(f'Correlation between ADR and Total Reservations: {correlation}')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure the date column is in datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], format='%d/%m/%Y')

# Extract month and year
df['month_year'] = df['reservation_status_date'].dt.to_period('M')

# Filter data for 2016 only
filtered_df_2016 = df[df['reservation_status_date'].dt.year == 2016]

# Further filter for cancelled and non-cancelled reservations in all hotels
cancelled_all_hotels = filtered_df_2016[filtered_df_2016['is_canceled'] == 1]
non_cancelled_all_hotels = filtered_df_2016[filtered_df_2016['is_canceled'] == 0]

# Group by month_year, then count cancellations and non-cancellations
monthly_cancellations_all_hotels = cancelled_all_hotels.groupby('month_year').size().reset_index(name='cancellations')
monthly_non_cancellations_all_hotels = non_cancelled_all_hotels.groupby('month_year').size().reset_index(name='non_cancellations')

# Merge the datasets on month_year
combined_data = pd.merge(monthly_cancellations_all_hotels, monthly_non_cancellations_all_hotels, on='month_year')

# Plot double bar graph
fig, ax = plt.subplots(figsize=(20, 10))
bar_width = 0.4

# Set positions for bars
r1 = range(len(combined_data))
r2 = [x + bar_width for x in r1]

# Plot bars
plt.bar(r1, combined_data['cancellations'], color='red', width=bar_width, label='Cancellations')
plt.bar(r2, combined_data['non_cancellations'], color='green', width=bar_width, label='Non-Cancellations')

# Add labels and title
plt.xlabel('Month-Year', fontsize=30)
plt.ylabel('Number of Reservations', fontsize=30)
plt.title('Cancellations and Non-Cancellations by Month (2016)', fontsize=30)
plt.xticks([r + bar_width/2 for r in range(len(combined_data))], combined_data['month_year'].astype(str), rotation=45)
plt.legend()
plt.show()
```
